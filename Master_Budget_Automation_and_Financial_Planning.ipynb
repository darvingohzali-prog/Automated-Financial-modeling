# **Sales Bugdeting**


### Business Scenario

**Texas Rex Apparel** is preparing its **Annual Master Budget** for the upcoming fiscal year. As the first step in the budgeting process, management must develop a **Sales Budget**, which serves as the foundation for all subsequent operating and financial budgets.

Based on market research, historical sales performance, and management's sales forecast, the company expects the following sales volume for the upcoming year:

| Quarter | Budgeted Unit Sales |
|:--------|--------------------:|
| Q1 | 1,000 |
| Q2 | 1,200 |
| Q3 | 1,500 |
| Q4 | 2,000 |

The company sells each customized T-shirt at a **selling price of $10 per unit**.

The Sales Budget will be used to estimate quarterly sales revenue and provide the basis for preparing the Production Budget, Cash Budget, and Budgeted Financial Statements.

---

## Required

Prepare the **Sales Budget** for each quarter and for the fiscal year.

```python

#Sales Budgeting
#Source Data
Sales_Unit = Budgeting.loc[Budgeting['Account'] == 'Sales Unit'].iloc[:,1:]
Price_Sell = Budgeting.loc[Budgeting['Account'] == 'Sell Prices'].iloc[:,1:]
#Calculation
Budgeted_Sales = Price_Sell.values * Sales_Unit.values
#Data Frame
Quarter = Budgeting.columns [1:]
df_Sales_Unit = pd.DataFrame(Sales_Unit.values, columns = Quarter, index = ['Unit'])
df_Price_Sell = pd.DataFrame(Price_Sell.values, columns = Quarter, index = ['Price Sell'])
df_Budgeted_Sales = pd.DataFrame(Budgeted_Sales, columns = Quarter, index = ['Budgeted Sales'])

#Table
df_Sales_Budget = pd.concat([df_Sales_Unit, df_Price_Sell, df_Budgeted_Sales])

#New Column year
df_Sales_Budget['year'] = df_Sales_Budget.apply(lambda row : row.mean() if row.name == 'Price Sell' else row.sum(), axis = 1)

#Title
styled_df = ( df_Sales_Budget.style
             .set_caption("Sales Budget")
            )
styled_df
```

# **Production Budget**

### Business Scenario

Following the completion of the **Sales Budget**, **Texas Rex Apparel** must determine the number of T-shirts to produce during each quarter of the upcoming fiscal year.

To ensure uninterrupted production while minimizing inventory holding costs, management has established an inventory policy requiring the company to maintain an ending inventory equal to **20% of the following quarter's expected unit sales**.

At the beginning of the budget year, the company has **180 finished T-shirts** in inventory. In addition, management forecasts that **first-quarter sales for the following fiscal year will be 1,000 units**, which will be used to determine the desired ending inventory for the fourth quarter.

The Production Budget will estimate the number of units that must be manufactured each quarter to satisfy customer demand while maintaining the company's inventory policy.

---

## Required

Using the information provided in the **Sales Budget**:

1. Calculate the **desired ending finished goods inventory** for each quarter and for the fiscal year.

2. Prepare the **Production Budget** for each quarter and for the fiscal year.

```python

#Production Budget

#Source
Beginning_Inventory = Budgeting.loc[Budgeting['Account'] == 'Beginning Inventory'].iloc[:,1:]
Ending_Inventory = Budgeting.loc[Budgeting['Account'] == 'Ending Inventory'].iloc[:,1:]

#Calculation

Total_Needs = Sales_Unit.values + Ending_Inventory.values
Unit_to_be_Produced = Total_Needs - Beginning_Inventory.values

#DataFrame

df_Beginning_Inventory = pd.DataFrame(Beginning_Inventory.values, columns = Quarter, index = ['Beginning Inventory'])
df_Ending_Inventory = pd.DataFrame(Ending_Inventory.values, columns = Quarter, index = ['Ending Inventory'])
df_Total_Needs = pd.DataFrame(Total_Needs, columns = Quarter, index = [' Total Needs'] )
df_Unit_to_be_Produced = pd.DataFrame(Unit_to_be_Produced, columns = Quarter, index = ['Unit to be Produced'])

#Table
df_Production_Budget = pd.concat([df_Sales_Unit,df_Beginning_Inventory,df_Total_Needs,df_Ending_Inventory,df_Unit_to_be_Produced])

#Year Column
df_Production_Budget['year'] = df_Sales_Budget.apply(lambda row : row.mean() if row.name == 'Price Sell' else row.sum(), axis = 1)

#Title
styled_df = ( df_Sales_Budget.style
             .set_caption("Production Budget")
            )
styled_df
```

# Manufacturing Overhead Budget



### Business Scenario

Following the completion of the **Direct Labor Budget**, **Texas Rex Apparel** must estimate its **Manufacturing Overhead Costs** for the upcoming fiscal year.

Manufacturing overhead consists of all indirect production costs that cannot be directly traced to individual T-shirts but are necessary to support the manufacturing process. These costs include both **variable** and **fixed** manufacturing overhead.

Management has established the following manufacturing overhead cost assumptions:

### Variable Manufacturing Overhead

- **$5.00 per direct labor hour**

### Fixed Manufacturing Overhead

- **$1,645 per quarter**

The fixed manufacturing overhead includes **$540 of depreciation expense** each quarter. Since depreciation is a **non-cash expense**, it should be identified separately for use in the **Cash Budget**.

The Manufacturing Overhead Budget will estimate total production overhead costs for each quarter and provide the information required to calculate product manufacturing costs and prepare subsequent budgets.

---

## Required

Using the **Direct Labor Budget** and the manufacturing overhead assumptions provided above:

1. Calculate the **variable manufacturing overhead** for each quarter.

2. Calculate the **total manufacturing overhead**, including both variable and fixed overhead costs.

3. Identify the **depreciation expense** included in manufacturing overhead for each quarter.

4. Prepare the **Manufacturing Overhead Budget** for each quarter and for the fiscal year.

```python
# --- OVERHEAD BUDGET ---

# Source Data
Variable_Overhead_Cost = Budgeting.loc[Budgeting['Account'] == 'Variable overhead cost rate'].iloc[:, 1:]
Fixed_Overhead_Cost = Budgeting.loc[Budgeting['Account'] == 'Budgeted Fixed overhead'].iloc[:, 1:]

# Calculation
# PERBAIKAN: Menggunakan Total_Hour_Needs dari perhitungan Direct Labor sebelumnya
Budgeted_Variable_Cost = Total_Hour_Needs * Variable_Overhead_Cost.values
Total_Overhead_Cost = Budgeted_Variable_Cost + Fixed_Overhead_Cost.values

# DataFrame
# Perhatikan kita memanggil df_Total_Hour_Needs yang sudah di-assign sebelumnya
df_Variable_Overhead_Cost = pd.DataFrame(Variable_Overhead_Cost.values, columns=Quarter, index=['Variable Overhead Cost Rate'])
df_Budgeted_Variable_Cost = pd.DataFrame(Budgeted_Variable_Cost, columns=Quarter, index=['Budgeted Variable Overhead'])
df_Fixed_Overhead_Cost = pd.DataFrame(Fixed_Overhead_Cost.values, columns=Quarter, index=['Budgeted Fixed Overhead'])
df_Total_Overhead_Cost = pd.DataFrame(Total_Overhead_Cost, columns=Quarter, index=['Total Overhead Cost'])

# Table Overhead Budget
# Menggabungkan df_Total_Hour_Needs di baris pertama agar visual outputnya persis seperti di Excel
df_Overhead_Budget = pd.concat([
    df_Total_Hour_Needs,
    df_Variable_Overhead_Cost,
    df_Budgeted_Variable_Cost,
    df_Fixed_Overhead_Cost,
    df_Total_Overhead_Cost
])

# Year Columns
# Logic tetap sama: Untuk rate kita ambil rata-rata (mean), sisanya dijumlahkan (sum)
df_Overhead_Budget['year'] = df_Overhead_Budget.apply(
    lambda row: row.mean() if row.name in ['Variable Overhead Cost Rate'] else row.sum(),
    axis=1
)

# Title
styled_df_Overhead = (
    df_Overhead_Budget.style
    .set_caption("Overhead Budget")
)

styled_df_Overhead
```

The COGS Budget will be used to estimate the total manufacturing costs associated with goods sold during the year. This budget is essential for preparing the Budgeted Income Statement and evaluating the company's expected gross profit.

The company reports a **beginning finished goods inventory** valued at **$1,251**.

The Cost of Goods Sold Budget should be prepared using information from the previously completed operating budgets.

---

## Required

Prepare the **Cost of Goods Sold (COGS) Budget** for each quarter and for the year

```python
# --- DIRECT MATERIAL USED BUDGET ---

# 1. Beginning Raw Material (Dari dataset)
# Pastikan nama akun sesuai dengan yang ada di csv ('Raw Material ')
Beginning_Raw_Material = Budgeting.loc[Budgeting['Account'] == 'Raw Material '].iloc[:,1].values[0]

# 2. Total Direct Material Purchased (Menggunakan perhitunganmu sebelumnya)
Total_Direct_Material_Purchased = df_Total_Direct_Material.loc['Total Direct Material', 'year']

# 3. Ending Direct Material (Persediaan Q4 dikali harga masing-masing bahan)
Ending_DM_A_Q4 = Budgeting.loc[Budgeting['Account'] == 'Ending Inventory Direct Material A'].iloc[:,-1].values[0]
Ending_DM_B_Q4 = Budgeting.loc[Budgeting['Account'] == 'Ending Inventory Direct Material B'].iloc[:,-1].values[0]
Price_DM_A = Budgeting.loc[Budgeting['Account'] == 'Direct material A cost (finish Good)'].iloc[:,1].values[0]
Price_DM_B = Budgeting.loc[Budgeting['Account'] == 'Direct material B cost (finish Good)'].iloc[:,1].values[0]

Ending_Direct_Material = (Ending_DM_A_Q4 * Price_DM_A) + (Ending_DM_B_Q4 * Price_DM_B)

# 4. Kalkulasi Direct Material Used
Direct_Material_Used = Beginning_Raw_Material + Total_Direct_Material_Purchased - Ending_Direct_Material

# DataFrame
year_col = ['year']
df_Direct_Material_Used = pd.DataFrame(
    [Beginning_Raw_Material, Total_Direct_Material_Purchased, Ending_Direct_Material, Direct_Material_Used],
    columns=year_col,
    index=['Beginning Raw Material', 'Direct Material Purchased', 'Ending Direct Material', 'Direct Material Used']
)

# Title
styled_df_DM_Used = (
    df_Direct_Material_Used.style
    .set_caption("Direct Material Used Budget")
)
styled_df_DM_Used
```

```python
# --- BUDGETED COST OF GOODS SOLD ---

# 1. Ekstrak nilai murni (float/scalar)
Direct_Labor_Used = float(df_Direct_Labor_Budget.loc['Total Direct Labor Cost', 'year'])
Overhead_Used = float(df_Overhead_Budget.loc['Total Overhead Cost', 'year'])
Total_Ending_Inventory_Val = float(df_Total_Ending_inventory.iloc[0, 0])

# 2. PERBAIKAN: Tarik Beginning Finished Goods dari dataset secara dinamis
# Menggunakan str.contains agar aman dari salah ketik atau extra spasi (' ') di Excel
Beginning_Finished_Goods = float(
    Budgeting.loc[Budgeting['Account'].str.contains('Finished Good', case=False, na=False)].iloc[:, 1].values[0]
)

# 3. Kalkulasi Akuntansi
Budgeted_Manufacturing_Cost = Direct_Material_Used + Direct_Labor_Used + Overhead_Used
Cost_of_Goods_Available_for_Sales = Budgeted_Manufacturing_Cost + Beginning_Finished_Goods
Cost_of_Good_Sold = Cost_of_Goods_Available_for_Sales - Total_Ending_Inventory_Val

# 4. Pembuatan DataFrame COGS
df_Budgeted_Cost_of_Good_Sold = pd.DataFrame(
    [
        Direct_Material_Used,
        Direct_Labor_Used,
        Overhead_Used,
        Budgeted_Manufacturing_Cost,
        Beginning_Finished_Goods,
        Cost_of_Goods_Available_for_Sales,
        Total_Ending_Inventory_Val,
        Cost_of_Good_Sold
    ],
    columns=['year'],
    index=[
        'Direct Material Used',
        'Direct Labor Used',
        'Overhead',
        'Budgeted Manufacturing Costs',
        'Beginning Finished Goods',
        'Cost of Goods Available for Sales',
        'Less : Ending Finished Goods',
        'Cost of Good Sold'
    ]
)

# 5. Visualisasi Tabel
styled_df_Budgeted_Cost_of_Good_Sold = (
    df_Budgeted_Cost_of_Good_Sold.style
    .set_caption("Budgeted Cost of Goods Sold")
)
styled_df_Budgeted_Cost_of_Good_Sold
```

# **CASH BUDGET**


As the final step in the operating budgeting process, **Texas Rex Apparel** is preparing its **Cash Budget** for the upcoming fiscal year. The Cash Budget will help management evaluate the company's liquidity position, determine financing requirements, and ensure that sufficient cash is available to support daily operations.

The Cash Budget should incorporate all projected cash inflows and outflows developed in the previous budgeting schedules, including cash collections from customers and cash payments for operating activities.

The company has established the following financial policies and assumptions:

### Financing Policy

- The company requires a **minimum cash balance of $1,000** at the end of each quarter.
- If the ending cash balance falls below this minimum, the company may borrow funds in **multiples of $1,000**.
- Borrowings are made **at the beginning of each quarter**.
- Loan repayments, including interest, are made **at the end of each quarter**.
- The annual interest rate on borrowings is **12%**, calculated only on the outstanding principal.

### Non-Cash Expenses

The following depreciation expenses are included in the operating budgets but **do not require cash payments**:

- Manufacturing Overhead Depreciation: **$540 per quarter**
- Selling and Administrative Depreciation: **$150 per quarter**

### Capital Expenditure

To expand production capacity, the company plans to purchase additional **screen-printing equipment** during the **first quarter**.

- Equipment Cost: **$6,500**
- The purchase will be paid **entirely in cash** during Q1.
- Management expects to finance this investment through a combination of operating cash flows and short-term borrowings if necessary.

### Income Taxes

- Estimated income tax payable for the year: **$3,469**
- The tax payment will be made **at the end of the fourth quarter**.

### Beginning Cash Balance

- Cash available at the beginning of the budget year: **$5,200**

### Additional Assumption

- All amounts presented in the budget are **rounded to the nearest dollar**.

---

## Required

Prepare the **Cash Budget** for each quarter and for the fiscal year.

```python
import numpy as np
import pandas as pd

# --- CASH BUDGET (ANTI-KEYERROR VERSION) ---

# 1. Ekstrak Data Aliran Kas Masuk (Inflows)
# Menggunakan .iloc[-1] untuk memaksa Python mengambil baris PALING BAWAH (Total)
# Ini membuat kode kebal terhadap salah ketik atau spasi tersembunyi di nama indeks
collections = df_Cash_Receive.iloc[-1, 0:4].values.flatten().astype(float)

# Tarik depresiasi dari dataset mentah
depr_oh = Budgeting.loc[Budgeting['Account'].str.contains('Depreciatin Overhead', case=False, na=False)].iloc[:, 1:5].values.flatten().astype(float)
depr_sa = Budgeting.loc[Budgeting['Account'].str.contains('Depreciation S&A', case=False, na=False)].iloc[:, 1:5].values.flatten().astype(float)

# 2. Ekstrak Data Aliran Kas Keluar (Disbursements)
disb_rm = np.array(Total_Cash_Needed, dtype=float)
disb_dl = df_Direct_Labor_Budget.loc['Total Direct Labor Cost'].iloc[0:4].values.astype(float)
disb_oh = df_Overhead_Budget.loc['Total Overhead Cost'].iloc[0:4].values.astype(float)
disb_sa = df_Budgeted_SA.loc['Total S&A Cost'].iloc[0:4].values.astype(float)

# PERBAIKAN: Menggunakan nama baris yang spesifik dan presisi
tax_val = float(df_Income_Statement_Budget.loc['Less: Income taxes'].iloc[-1])
taxes = np.array([0.0, 0.0, 0.0, tax_val])

# Tarik pengeluaran modal (Equipment) secara dinamis
equip_val = float(Budgeting.loc[Budgeting['Account'].str.contains('New Building', case=False, na=False)].iloc[:, 1].values[0])
equip = np.array([equip_val, 0.0, 0.0, 0.0])

# 3. Ekstrak Data Financing
borrow_val = float(Budgeting.loc[Budgeting['Account'].str.contains('Borrowing', case=False, na=False)].iloc[:, 1].values[0])
borrowing = np.array([borrow_val, 0.0, 0.0, 0.0])

repayment = np.array([0.0, -borrow_val, 0.0, 0.0])

interest_val = float(Budgeting.loc[Budgeting['Account'].str.contains('Interest Expense', case=False, na=False)].iloc[:, 1].values[0])
interest = np.array([0.0, -interest_val, 0.0, 0.0])

total_financing = borrowing + repayment + interest

# 4. Inisiasi Array untuk Perhitungan Iteratif
beg_cash = np.zeros(4)
beg_cash[0] = 5200.0  # Saldo awal kas Q1

tot_cash_avail = np.zeros(4)
tot_disb = np.zeros(4)
excess_cash = np.zeros(4)
end_cash = np.zeros(4)

# 5. Iterasi Kalkulasi per Kuartal (Cascading Cash Balance)
for i in range(4):
    tot_cash_avail[i] = beg_cash[i] + collections[i] + depr_oh[i] + depr_sa[i]
    tot_disb[i] = -(disb_rm[i] + disb_dl[i] + disb_oh[i] + disb_sa[i] + taxes[i] + equip[i])
    excess_cash[i] = tot_cash_avail[i] + tot_disb[i]
    end_cash[i] = excess_cash[i] + total_financing[i]

    # Passing saldo akhir ke saldo awal kuartal berikutnya
    if i < 3:
        beg_cash[i+1] = end_cash[i]

# 6. Pembuatan DataFrame Cash Budget
Quarter_Cols = ['1', '2', '3', '4']

df_Cash_Budget = pd.DataFrame(
    [
        beg_cash, collections, depr_oh, depr_sa, tot_cash_avail,
        -disb_rm, -disb_dl, -disb_oh, -disb_sa, -taxes, -equip, tot_disb,
        excess_cash, borrowing, repayment, interest, total_financing, end_cash
    ],
    columns=Quarter_Cols,
    index=[
        'Beginning Cash Balance', 'Cash Sales and Collection on Account', 'Depreciation Overhead', 'Depreciation S&A', 'Total Cash available',
        'Raw Materials', 'Direct labor', 'Overhead', 'Selling and Administrative expense', 'Income Taxes', 'Equipment', 'Total Disbursment',
        'Excess /Deficiency Cash available over needs', 'Borrowing', 'Repayment', 'Interest', 'Total Financing', 'Ending Cash Balance'
    ]
)

# 7. Penyesuaian Kolom 'year' untuk Saldo Kumulatif Akuntansi
df_Cash_Budget['year'] = df_Cash_Budget.apply(
    lambda row:
        row.iloc[0] if row.name == 'Beginning Cash Balance' else
        row.iloc[3] if row.name == 'Ending Cash Balance' else
        row.sum(),
    axis=1
)

# --- PERBAIKAN: Membulatkan seluruh angka ke bilangan bulat ---
df_Cash_Budget = df_Cash_Budget.round(0)

# 8. Visualisasi Tabel dengan Formatting Akuntansi
styled_df_Cash_Budget = (
    df_Cash_Budget.style
    .format("{:,.0f}") # Memastikan pemisah ribuan tampil tanpa angka desimal
    .set_caption("Cash Budget")
)

styled_df_Cash_Budget
```

```python
import pandas as pd

# ==========================================
# ALTMAN Z-SCORE (PREDICTIVE BANKRUPTCY MODEL)
# ==========================================

# 1. Menyiapkan Komponen Data dari Budget Sebelumnya
# Working Capital = Current Assets - Current Liabilities
working_capital = total_current_assets - current_liabilities
total_assets_z = total_assets
retained_earnings_z = re_end           # Diambil dari Saldo Laba Neraca
ebit_z = op_income                     # Laba Operasional (Operating Income)
book_value_equity_z = total_equity     # Sebagai proxy untuk Market Value of Equity
total_liabilities_z = total_liabilities
sales_z = sales

# 2. Menghitung 5 Rasio Pembentuk Z-Score (X1 - X5)
X1 = working_capital / total_assets_z
X2 = retained_earnings_z / total_assets_z
X3 = ebit_z / total_assets_z
X4 = book_value_equity_z / total_liabilities_z
X5 = sales_z / total_assets_z

# 3. Kalkulasi Altman Z-Score (Klasik / Original Model)
z_score = (1.2 * X1) + (1.4 * X2) + (3.3 * X3) + (0.6 * X4) + (1.0 * X5)

# 4. Klasifikasi Zona Kesehatan Finansial
if z_score > 2.99:
    zone = "Safe Zone (Finansial Sangat Sehat)"
elif 1.81 <= z_score <= 2.99:
    zone = "Grey Zone (Area Waspada)"
else:
    zone = "Distress Zone (Risiko Kebangkrutan Tinggi)"

# 5. Susun ke dalam DataFrame
df_Z_Score = pd.DataFrame(
    [X1, X2, X3, X4, X5, z_score],
    columns=['Result'],
    index=[
        'X1: Working Capital / Total Assets',
        'X2: Retained Earnings / Total Assets',
        'X3: EBIT / Total Assets',
        'X4: Book Value of Equity / Total Liabilities',
        'X5: Sales / Total Assets',
        'Altman Z-Score Final'
    ]
)

# 6. Visualisasi Tabel
styled_Z_Score = (
    df_Z_Score.style
    .format("{:.4f}") # Z-Score sangat sensitif, disarankan pakai 4 angka di belakang koma
    .set_caption(f"Altman Z-Score Analysis | Status: {zone}")
    .set_table_styles([{
        'selector': 'caption',
        'props': [('font-size', '14px'), ('font-weight', 'bold')]
    }])
)

display(styled_Z_Score)
```

```python
import pandas as pd
import numpy as np
import os
from openpyxl.styles import Font, Alignment, numbers

# Asumsi bahwa DataFrames berikut sudah terdefinisi di lingkungan Anda:
# df_Sales_Budget, df_Production_Budget, df_Material_A_Budget, df_Direct_Material_B_Budget,
# df_Direct_Labor_Budget, df_Overhead_Budget, df_Budgeted_SA, df_Income_Statement_Budget,
# df_Cash_Receive, df_Cash_Payment, df_Cash_Budget, df_SOFP, df_Financial_Analysis, df_Z_Score

# Daftar tabel yang akan diekspor, disesuaikan dengan variabel global yang ada.
# Saya telah menambahkan Material B dan memisahkan Financial Analysis menjadi dua bagian.
semua_tabel = [
    ("1. Sales Budget", df_Sales_Budget),
    ("2. Production Budget", df_Production_Budget),
    ("3. Direct Material Budget (Material A)", df_Direct_Material_A_Budget),
    ("4. Direct Material Budget (Material B)", df_Direct_Material_B_Budget),
    ("5. Direct Labor Budget", df_Direct_Labor_Budget),
    ("6. Manufacturing Overhead Budget", df_Overhead_Budget),
    ("7. Selling & Admin Budget", df_Budgeted_SA), # Menggunakan df_Budgeted_SA (variabel global)
    ("8. Budgeted Income Statement", df_Income_Statement_Budget), # Menggunakan df_Income_Statement_Budget (variabel global)
    ("9. Schedule of Cash Receipts", df_Cash_Receive), # Menggunakan df_Cash_Receive (variabel global)
    ("10. Schedule of Cash Payments", df_Cash_Payment),
    ("11. Cash Budget", df_Cash_Budget),
    ("12. Budgeted SOFP (Neraca)", df_SOFP),
    ("13. Financial Analysis", df_Financial_Analysis), # Menambahkan Financial Analysis
    ("14. Altman Z-Score Analysis", df_Z_Score)
]

output_excel_filename = 'Master_Budget_Report_Single_Sheet.xlsx'

print("Mendesain file Excel (Satu Sheet Lengkap) sesuai permintaan...")

with pd.ExcelWriter(output_excel_filename, engine='openpyxl') as writer:
    workbook = writer.book
    worksheet = workbook.create_sheet('Master_Budget_Report')
    writer.sheets['Master_Budget_Report'] = worksheet
    worksheet.sheet_view.showGridLines = False

    current_row = 1
    for title, df in semua_tabel:
        # Tulis Judul Tabel
        worksheet.cell(row=current_row, column=1, value=title).font = Font(bold=True, size=12)
        current_row += 1 # Pindah ke baris berikutnya untuk DataFrame

        # Tentukan label indeks untuk DataFrame
        index_label = 'Account / Indicator' if title.startswith(('12.', '13.', '14.')) else 'Account'

        # Tulis DataFrame ke lembar kerja. Header DataFrame akan dimulai dari `current_row`.
        df.to_excel(writer, sheet_name='Master_Budget_Report', startrow=current_row, startcol=0, index_label=index_label)

        # Terapkan format tebal pada baris header DataFrame (yaitu baris `current_row`)
        for col_idx in range(1, len(df.columns) + 2): # Meliputi kolom indeks dan semua kolom data
            worksheet.cell(row=current_row, column=col_idx).font = Font(bold=True)

        # Pindahkan kursor baris untuk tabel berikutnya
        # current_row + len(df) + 1 (untuk baris header DF + baris data DF)
        # + 3 (untuk 3 baris kosong antar tabel)
        current_row += len(df) + 1 + 3

    # Rapikan lebar kolom
    worksheet.column_dimensions['A'].width = 45
    for col_letter in ['B', 'C', 'D', 'E', 'F', 'G']:
        worksheet.column_dimensions[col_letter].width = 15

    # Formatting Angka (Koma ribuan dan Desimal)
    # Iterasi dimulai dari baris pertama (Excel) dan kolom B hingga G
    for row in worksheet.iter_rows(min_row=1, max_row=worksheet.max_row, min_col=2, max_col=7):
        for cell in row:
            if isinstance(cell.value, (int, float)):
                # Jika angka memiliki nilai absolut kecil (misal: rasio desimal) DAN memiliki desimal,
                # format dengan 3 angka desimal. Selain itu, gunakan pemisah ribuan tanpa desimal.
                if abs(cell.value) < 100 and (cell.value % 1 != 0):
                    cell.number_format = '0.000'
                else:
                    cell.number_format = numbers.FORMAT_NUMBER_COMMA_SEPARATED1

    # Hapus default 'Sheet' bawaan openpyxl jika ada
    if 'Sheet' in workbook.sheetnames:
        workbook.remove(workbook['Sheet'])

print(f"✅ BERHASIL! File telah direplikasi ke dalam satu Sheet.")
print(f"📂 Nama File: {os.path.abspath(output_excel_filename)}")
```
